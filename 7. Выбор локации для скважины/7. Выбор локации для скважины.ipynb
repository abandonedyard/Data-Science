## 2. Обучение и проверка модели

```python
# Создадим списки с целевыми признаками валидационной выборки
# Разобьем данные на обучающую и валидационную выборки в соотношении 75:25
# Обучим модель и сделаем предсказания на валидационной выборке
# Сохраним их и RMSE в заготовленных списках
# Создадим таблицу по этим спискам

target_list = []
prediction_list = []
rmse_list = []
region_number_list = []



for i, j in zip(data_list, range(1, 4)):
    X = (i.drop(['id', 'product'], axis=1))
    y = (i['product'])
    X_train, X_valid, y_train, y_valid = (
    train_test_split(X, y, train_size=0.75, 
                     test_size=0.25, random_state=12345)
    )
    target_list.extend(y_valid)
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    prediction_list.extend(predictions)
    rmse_list.extend(
        [mean_squared_error(y_valid, predictions) ** 0.5] * len(y_valid)
    )
    region_number_list.extend([j] * len(y_valid))
    
    
prediction_data = pd.DataFrame(
        data={'region_number': region_number_list, 'target': target_list, 
              'prediction': prediction_list, 'rmse': rmse_list}
        )
```

```python
# Напечатайте на экране средний запас предсказанного сырья и RMSE модели

print('Средний запас предсказанного сырья и RMSE модели по каждому региону')

prediction_data.pivot_table(
    index = ['region_number'], 
    values = ['prediction', 'rmse']
)
```

###  Вывод обучения и проверка модели
        
       
Исходя из квадрата средней квадротичной ошибки каждого региона, модель отработала очень плохо на 1-ом и 3-ем регионе, но очень хорошо на 2-ом. Однако, как мы уже заметили при подготовке данных, во втором регионе всего 12 уникальных значений в столбце целевого признака, поэтому обучаться ей было легко.

## 3. Подготовка к расчёту прибыли

```python
# Объявим переменные для расчета прибыли

total_wells = 500 # При разведке региона исследуют 500 точек
best_wells = 200  # 200 лучших для разработки
budget = 10000000000 # Бюджет на разработку скважин в регионе — 10 млрд рублей
revenue_per_barrel = 450 # Один баррель сырья приносит 450 рублей дохода
revenue_per_unit = 450000 # Доход с каждой единицы продукта составляет 450 тыс. рублей
```

```python
# Рассчитаем достаточный объём сырья для безубыточной разработки новой скважины

volume_for_breakeven = budget / best_wells / revenue_per_unit
print(f'Для безубыточной разработки новой скважины необходимо {round(volume_for_breakeven, 2)} тысяч баррелей сырья')
```

```python
# Посмотрим средний запас сырья в каждом регионе
# Посчитаем его процент от необходимого прога безубыточности

for i, j in zip(data_list, range(1, 4)):
    volume = round(i['product'].mean(), 2)
    
    print(
        f'Средний запас сырья в регионе №{j}: {volume} тысяч баррелей. '
        f'Что состоавляет {volume / volume_for_breakeven :.0%} от необходимого объема.'
    )
```

###  Вывод подготовки к расчёту прибыли
        
       
Ни один из регионов в среднем не проходит границу безубыточности. Особенно плохи дела во втором регионе – на 38% меньше необходимого.

## 4. Напишем функцию для расчёта прибыли по выбранным скважинам и предсказаниям модели

```python
# Функция выберает скважины с максимальными значениями предсказаний
# Суммирует целевое значение объёма сырья, соответствующее этим предсказаниям
# Возвращает прибыль для полученного объёма сырья

def revenue(target, probabilities):
    probs_sorted = probabilities.sort_values(ascending=False)
    selected = target[probs_sorted.index][:best_wells] 
    return revenue_per_unit * selected.sum() - budget
```

## 5. Расчёт прибыли и рисков 

```python
# Применим технику Bootstrap с 1000 выборок, чтобы найти распределение прибыли

state = np.random.RandomState(12345)

values_all = []

for i in range(1, 4):
    values = []
    for j in range(1000):
        target_subsample = (
            prediction_data[prediction_data['region_number'] == i]['target']
            .sample(n=500, replace=True, random_state=state)
        )
        probs_subsample = (
            prediction_data['prediction'][target_subsample.index]
        )
        values.append(revenue(target_subsample, probs_subsample))
    values_all.append(values)
```

```python
# Найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков

for i, j in zip(values_all, range(1, 4)):
    print(f'Регион №{j}')
    print()
    values = pd.Series(i) 
    profit = round(values.mean())
    print(f'Средняя прибыль: {profit} руб.' )
    
    confidence_interval = st.t.interval(
    0.95, len(values)-1, values.mean(), values.sem()
    ) 
    print(f'95%-ый доверительный интервал: {confidence_interval}')
    risk = round((values<0).sum()*0.1, 1)
    print(f'Риск убытков: {risk}%')
    print()
    print()
```
